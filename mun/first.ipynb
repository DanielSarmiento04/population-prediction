```python
import pandas as pd
```

## Load data from dane (xls)

```python
path = "ProyeccionMunicipios2005_2020.xls"
sheetName = "Mpios"
```

### Note that the first 8 rows of the document are not necessarily, it is clean data process.

```python
df = pd.read_excel(path, sheet_name=sheetName, skiprows=8)
```

```python
df.head()
```

### get the specific municipality data

```python
gramalote_row = df.loc[df['MPIO'] == 'Gramalote']
```

```python
gramalote_row.head()
```

### Convert to efficient file manipulation

```python
# Save the row to a pickle file
gramalote_row.to_pickle("gramalote_row.pkl", compression="gzip")
```
